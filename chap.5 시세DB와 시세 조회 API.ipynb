```python
#책에서는 야후 파이낸스 데이터의 문제점을 액면분할 이전과 이후의 차이를 보이지 않는다고 하지만 지금은 액면 분할이 반영되어있는 것을 확인함
#하지만 중간중간 volume 데이터가 몇주 단위로 비어있는 것은 확인할 수 있음
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
import matplotlib.pyplot as plt

df = pdr.get_data_yahoo('005930.KS', '2017-01-01')

plt.figure(figsize = (9,6))
plt.subplot(2,1,1)
plt.title('samsung (by yahoo)')
plt.plot(df.index, df['Close'], 'c', label = 'Close')
plt.plot(df.index, df['Adj Close'], 'b--',label = 'Adj Close')
plt.legend(loc = 'best')
plt.subplot(2,1,2)
plt.bar(df.index, df['Volume'], color = 'g', label = 'Volume')
plt.legend(loc = 'best')
plt.show()
```
