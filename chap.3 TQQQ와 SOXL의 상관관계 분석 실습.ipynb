```python
#matplotlib 한글 깨짐 문제 설정
from matplotlib import rc
import matplotlib.pyplot as plt
rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False
```

```python
from pandas_datareader import data as pdr
import yfinance as yf
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
yf.pdr_override()

tqqq = pdr.get_data_yahoo('TQQQ', '2013-01-01')
soxl = pdr.get_data_yahoo('SOXL','2013-01-01')

plt.figure(figsize = (9,5))
plt.title('TQQQ와 SOXL의 종가 변화량')
plt.plot(tqqq.index, tqqq.Close, 'r--', label = 'TQQQ')
plt.plot(soxl.index, soxl.Close, 'b', label = 'SOXL')
plt.grid(True)
plt.legend(loc = 'best')
plt.show()
```

```python
regr = stats.linregress(tqqq.Close, soxl.Close)
plt.scatter(tqqq.Close, soxl.Close, marker = '.')
plt.title(f'TQQQ X SOXL (R = {regr.rvalue:.2f})')
plt.plot(tqqq.Close, regr.slope * tqqq.Close + regr.intercept, 'r')
plt.xlabel("TQQQ")
plt.ylabel("SOXL")
```
